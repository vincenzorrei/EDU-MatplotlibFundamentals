```python
# Working with Subplots
# This script demonstrates how to create and customize multiple subplots

import matplotlib.pyplot as plt
import numpy as np

# Generate sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = x**2
y5 = np.exp(x/10)

# Method 1: subplot() function
plt.figure(figsize=(12, 10))

plt.subplot(2, 2, 1)  # 2 rows, 2 columns, 1st subplot
plt.plot(x, y1)
plt.title('Sine Function')
plt.grid(True)

plt.subplot(2, 2, 2)  # 2 rows, 2 columns, 2nd subplot
plt.plot(x, y2)
plt.title('Cosine Function')
plt.grid(True)

plt.subplot(2, 2, 3)  # 2 rows, 2 columns, 3rd subplot
plt.plot(x, y4)
plt.title('Quadratic Function')
plt.grid(True)

plt.subplot(2, 2, 4)  # 2 rows, 2 columns, 4th subplot
plt.plot(x, y5)
plt.title('Exponential Function')
plt.grid(True)

plt.tight_layout()  # Adjust spacing between subplots
plt.savefig('subplot_method1.png')
plt.show()

# Method 2: subplots() function
fig, axs = plt.subplots(2, 3, figsize=(15, 8))

# First row
axs[0, 0].plot(x, y1, 'r-')
axs[0, 0].set_title('Sine Function')

axs[0, 1].plot(x, y2, 'g-')
axs[0, 1].set_title('Cosine Function')

axs[0, 2].plot(x, y3, 'b-')
axs[0, 2].set_title('Tangent Function')

# Second row
axs[1, 0].plot(x, y4, 'm-')
axs[1, 0].set_title('Quadratic Function')

axs[1, 1].plot(x, y5, 'c-')
axs[1, 1].set_title('Exponential Function')

# Hide the empty subplot
axs[1, 2].axis('off')

# Add a main title
fig.suptitle('Multiple Mathematical Functions', fontsize=16)

# Improve spacing
fig.tight_layout()
plt.subplots_adjust(top=0.9)  # Adjust to make room for the suptitle

plt.savefig('subplot_method2.png')
plt.show()
```
