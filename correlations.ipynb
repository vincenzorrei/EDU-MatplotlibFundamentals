```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
credit_df = pd.read_csv('https://raw.githubusercontent.com/vincenzorrei/EDU-Datasets/refs/heads/main/Credit.csv')

credit_df.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
corr_matrix = credit_df.select_dtypes(include=["number"]).corr()
corr_matrix
```

```python
# Creiamo la heatmap con parametri personalizzati
plt.figure(figsize=(10, 8))
sns.heatmap(
    corr_matrix,
    cmap="magma_r",
    annot=True,
    fmt=".2f",
    linewidths=5,
    linecolor="white",
    vmin=-1,
    vmax=1,
    cbar_kws={"shrink": 0.8, "orientation": "vertical"},  # Personalizza la barra dei colori
    square=True
)

plt.title("Heatmap con maschera e personalizzazioni")
plt.show()
```

```python
# Seaborn pairplot
sns.pairplot(credit_df, kind='scatter', diag_kind='kde')
```
